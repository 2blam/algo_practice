You are given the heads of two sorted linked lists ```list1``` and ```list2```.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

### Example 1
```
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
```

### Example 2:
```
Input: list1 = [], list2 = []
Output: []
```

### Example 3:
```
Input: list1 = [], list2 = [0]
Output: [0]
```

````python
#Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def mergeTwoLists(self, list1, list2):
        dummy_head = ListNode()
        r = dummy_head
        l1, l2 = list1, list2
        while l1 and l2:
            if l1.val <= l2.val:
                r.next = l1
                l1 = l1.next
            else:
                r.next = l2
                l2 = l2.next
            r = r.next
            
        # tidy
        if l1 is not None:
            r.next = l1
        if l2 is not None:
            r.next = l2

        return dummy_head.next

l1 = ListNode(1)
l1.next = ListNode(2)
l1.next.next = ListNode(4)

l2 = ListNode(1)
l2.next = ListNode(3)
l2.next.next = ListNode(4)

r = Solution().mergeTwoLists(l1, l2)

n = r
while n:
    print(n.val, end=">")
    n = n.next

print()


l1 = None
l2 = None

r = Solution().mergeTwoLists(l1, l2)

n = r
while n:
    print(n.val, end=">")
    n = n.next

print()

l1 = None
l2 = ListNode(0)

r = Solution().mergeTwoLists(l1, l2)

n = r
while n:
    print(n.val, end=">")
    n = n.next

print()

# Time Complexity: O(n + m): n - len of list 1; m - len of list 2
# Space Complexity: O(1)
````
