You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

Example 1:

- 1 > 2 > 4
- 1 > 3 > 4

1 > 1 > 2 > 3 > 4 > 4

```
Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
```

Example 2:
```
Input: list1 = [], list2 = []
Output: []
```

Example 3:
```
Input: list1 = [], list2 = [0]
Output: [0]
```

Constraints:

- The number of nodes in both lists is in the range [0, 50].
- -100 <= Node.val <= 100
- Both list1 and list2 are sorted in non-decreasing order.

````python
# time complexity: O(n+m); n - num of nodes in list1; m - num of nodes in list2
# space complexity: O(1); only changing the pointers between list1 and list2
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
        
def print_linked_list(ListNode):
    while ListNode is not None:
        print(ListNode.val, end="->")
        ListNode = ListNode.next
        
def merge_two_sorted_list(list1, list2):
    before_head = ListNode(-1) # create a dummy node
    prev = before_head
    while list1 and list2:
        if (list1.val < list2.val):
            prev.next = list1
            prev = prev.next
            list1 = list1.next
        else:
            prev.next = list2
            prev = prev.next
            list2 = list2.next
    # checking
    prev.next = list1 if list1 is not None else list2
    return before_head.next

list1 = ListNode(1)
list1.next = ListNode(2)
list1.next.next = ListNode(4)

list2 = ListNode(1)
list2.next = ListNode(3)
list2.next.next = ListNode(4)

print_linked_list(list1)
print()
print_linked_list(list2)
head = merge_two_sorted_list(list1, list2)
print()
print_linked_list(head)
print()
print("*" * 10)

list1 = None
list2 = None
print_linked_list(list1)
print()
print_linked_list(list2)
head = merge_two_sorted_list(list1, list2)
print()
print_linked_list(head)
print("*" * 10)

list1 = None
list2 = ListNode(0)
print_linked_list(list1)
print()
print_linked_list(list2)
head = merge_two_sorted_list(list1, list2)
print()
print_linked_list(head)
        
````

````python
# time complexity: O(n + m)
# space complexity: O(n + m)
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
        
def print_linked_list(ListNode):
    while ListNode is not None:
        print(ListNode.val, end="->")
        ListNode = ListNode.next

def merge_two_sorted_list(list1, list2):
    curr1 = list1
    curr2 = list2
    head, curr = None, None
    
    while curr1 is not None and curr2 is not None:
        if curr1 and curr2:
            v1 = curr1.val
            v2 = curr2.val
            if v1 < v2:
                if head is None:
                    head = ListNode(v1)
                    curr = head
                else:
                    curr.next = ListNode(v1)
                    curr = curr.next
                curr1 = curr1.next
            else:
                if head is None:
                    head = ListNode(v2)
                    curr = head
                else:
                    curr.next = ListNode(v2)
                    curr = curr.next
                curr2 = curr2.next
    # check the remaining
    while(curr1 is not None):
        if curr is None:
            head = ListNode(curr1.val)
            curr = head
        else:
            curr.next = ListNode(curr1.val)
            curr = curr.next
        curr1 = curr1.next
    while(curr2 is not None):
        if curr is None:
            head = ListNode(curr2.val)
            curr = head
        else:
            curr.next = ListNode(curr2.val)
            curr = curr.next
        curr2 = curr2.next
    
    return head

list1 = ListNode(1)
list1.next = ListNode(2)
list1.next.next = ListNode(4)

list2 = ListNode(1)
list2.next = ListNode(3)
list2.next.next = ListNode(4)

print_linked_list(list1)
print()
print_linked_list(list2)
head = merge_two_sorted_list(list1, list2)
print()
print_linked_list(head)
print()
print("*" * 10)

list1 = None
list2 = None
print_linked_list(list1)
print()
print_linked_list(list2)
head = merge_two_sorted_list(list1, list2)
print()
print_linked_list(head)
print("*" * 10)

list1 = None
list2 = ListNode(0)
print_linked_list(list1)
print()
print_linked_list(list2)
head = merge_two_sorted_list(list1, list2)
print()
print_linked_list(head)
````
