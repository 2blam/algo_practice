Given the root of a binary tree, invert the tree, and return its root.

Example 1:
```
        4
       / \
      2   7
     / \ / \
    1  3 6  9
    
 =>
        4
       / \
      7   2
     / \ / \
    9  6 3  1
 
Input: root = [4,2,7,1,3,6,9]
Output: [4,7,2,9,6,3,1]
 ```
 
 Example 2:
 ```
    2
   / \
  1   3
  
 =>
 
    2
   / \
  3   1
  
Input: root = [2,1,3]
Output: [2,3,1]
 ```
 
 Example 3:
 ```
 Input: root = []
 Output: []
 ```

````python
# time complexity: O(n)
# space complexity: O(d) - depth of the tree
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def invert_binary_tree_helper(head):
    if head == None:
        return None
    right = invert_binary_tree_helper(head.right)
    left = invert_binary_tree_helper(head.left)
    head.right = left
    head.left = right
    return head

    
def invert_binary_tree(node):
    return invert_binary_tree_helper(node)
    
    
# example 1
e1 = TreeNode(4)
e1.left = TreeNode(2)
e1.right = TreeNode(7)
e1.left.left = TreeNode(1)
e1.left.right = TreeNode(3)
e1.right.left = TreeNode(6)
e1.right.right = TreeNode(9)

# example 2
e2 = TreeNode(2)
e2.left = TreeNode(1)
e2.right = TreeNode(3)

# example 3
e3 = None



it = invert_binary_tree(e1)
print(it.val)
print(it.left.val, it.right.val)
print(it.left.left.val, it.left.right.val, it.right.left.val, it.right.right.val)

print()
it = invert_binary_tree(e2)
print(it.val)
print(it.left.val, it.right.val)
````
